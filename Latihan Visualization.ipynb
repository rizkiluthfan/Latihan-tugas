```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
happy_2019 = pd.read_csv('happy_2019.csv')
```

```python
happy_2019
```

## Bubble Plot

```python
plt.figure(figsize = (10,10))

plt.scatter(x = happy_2019['Ladder'], y = happy_2019['Positive affect'], 
            s = happy_2019['Negative affect'], alpha = 0.5)

plt.xlabel('Ladder')
plt.ylabel('Positive affect')
plt.title('Bubble Plot')
plt.show()
```

```python
fig = plt.figure(figsize = (10,10))

plt.scatter(x = happy_2019['Ladder'], y = happy_2019['Positive affect'], 
            s = happy_2019['Negative affect'], alpha = 0.5, c = happy_2019['Social support']) #c = color

plt.xlabel('Ladder')
plt.ylabel('Positive Affect')
plt.title('Bubble Plot')

# menambah warna

ax_col = fig.add_axes([0.99, 0.3, 0.05, 0.3])
ax_col.set_title('Social Support Color Bar')
plt.colorbar(cax = ax_col)
plt.show()
```

## Boxplot

```python
plt.figure(figsize = (12,12))

sns.boxplot(data = happy_2019, x = 'Ladder', y= 'Region')
sns.swarmplot(data = happy_2019, x = 'Ladder', y= 'Region')
```

```python
plt.figure(figsize = (10,10))
sns.heatmap(happy_2019.corr()[['Ladder']], annot = True)
```

From data on heatmap above, we can see that many factors seems to correlate and might affect the Ladder score. First there is Social Support with 0.82 points, then there is Log of GDP Per Capita and Healthy Life Expectancy features with 0.81 points respectively. This might indicates that there is a strong posistive connection between those variables. While SD of Ladder, Positive affect, Negative affect and Freedom might also have a correlation, yet from the score above we might not see a strong connections as they only have a values of 0.52 to 0.55.
