Rather than 'Landsize', after some considerations I concluded that 'Price' category is making a lot more senses to fill the Missing Values(NaN) in the Car category. Beacuse after digging deep through the datas, I still found some confusing points, therefore I'd rather not to use it to fill the NaN in the Car category.  
That being said, now we'll star to replace the NaN values, categorized by Price.

```python
cat1 = melbourne['Car'][(melbourne['Price']>=85000) & (melbourne['Price']<=650000)]
cat2 = melbourne['Car'][(melbourne['Price']>=650001) & (melbourne['Price']<=901000)]
cat3 = melbourne['Car'][(melbourne['Price']>=901001) & (melbourne['Price']<=1328000)]
cat4 = melbourne['Car'][(melbourne['Price']>=1328001) & (melbourne['Price']<=2345000)]
cat5 = melbourne['Car'][(melbourne['Price']>2345000)]
```

```python
cat1 = cat1.fillna(1)
cat2 = cat2.fillna(1)
cat3 = cat3.fillna(2)
cat4 = cat4.fillna(2)
cat5 = cat5.fillna(2)
```

```python
filled_cars = pd.concat([cat1, cat2, cat3, cat4, cat5])
```

```python
melbourne['Car'] = melbourne['Car'].fillna(filled_cars)
```
