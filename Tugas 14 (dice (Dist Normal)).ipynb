```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import statistics as sts
%matplotlib inline
```

```python
def dice_roll(dice, n_roll):
    
    dsum = []
    for i in range(n_roll) :
        d = []
        for j in range(dice):
            d.append(random.randint(1,7))
            dsum.append(sum(d))
        return sns.distplot(dsum)
```
