```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

import warnings
```

```python
df_canc = pd.read_csv('cancer_data.csv')
```

```python
df_canc.drop(['Unnamed: 32', 'id'], axis = 1, inplace = True)
```

```python
df_canc.diagnosis = [1 if each == 'M' else 0 for each in df_canc.diagnosis]
```

```python
df_canc.head()
```

### Split Dataset

```python
x = df_canc.drop(['diagnosis'], axis = 1)
y = df_canc['diagnosis']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, shuffle = False)
```

### Fitting Model

```python
model = LogisticRegression(solver = 'lbfgs')
model.fit(x_train, y_train)
print(model.coef_)
print(model.intercept_)

m = model.coef_[0][0]
c = model.intercept_[0]
```

### Predict

```python
y_pred = model.predict(x_test)
```

### Confusion Matrix

```python
confusion_matrix(y_test, y_pred)
```

```python
fig, ax = plt.subplots(figsize = (8,5))
sns.heatmap(confusion_matrix(y_test, y_pred), annot = True, fmt = '.0f', ax = ax)
plt.xlabel('y aktual')
plt.ylabel('y prediksi')
plt.show()
```

TP (True Positive)  = Benar nebak positif (actual = YES, pred = YES)

TN (True Negative)  = Benar nebak negatif (actual = NO, pred = NO)

FP (False Positive) = Salah nebak positif (actual = NO, pred = YES)

FN (False Negative) = Salah nebak negatif (actual = YES, pred = NO)

```python
#confusion matrix report
print(classification_report(y_test, y_pred))
```
